````python
class Animal:
    def __init__(self, name, sound, category):
        self.name = name
        self.sound = sound
        self.category = category

    def make_sound(self):
        return f"{self.name} says {self.sound}!"
    
    def get_catagory(self):
        return self.category
    
    class Mammal(Animal):
        def __init__(self, name, sound):
            super().__init__(name, sound, "Mammal")

    class Bird(Animal):
        def __init__(self, name, sound):
            super().__init__(name, sound, "Bird")

    class Reptile(Animal):
        def __init__(self, name, sound):
            super().__init__(name, sound, "Reptile")

# Animals on McDonalds Farm
cow = Mammal("Cow", "Moo")
pig = Mammal("Pig", "Oink")
chicken = Bird("Chicken", "Cluck")
duck = Bird("Duck", "Quack")
snake = Reptile("Snake", "Hiss")

#Grouping Animals

farm_animals = [cow, pig, chicken, duck, snake]
catagorized_animals = {}

for animal in farm_animals:
    category = animal.get_category()
    if category not in categorized_animals:
        categorized_animals[category] = []
    categorized_animals[category].append(animal.name)

#Display animals
for category, animals in categorized_animals.items():
    print(f"{category}s: {', '.join(animals)}")
    
````
